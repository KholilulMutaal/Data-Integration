```python
tugas1=pd.read_csv("D:\Documents\Kuliah\SEMESTER 6\DATA MINING\Tugas 3 (Transformasi Data)\Concrete_Data.csv")
tugas1['Cement']=tugas1['Cement'].fillna((tugas1['Cement'].mean()))
tugas1['Blast Furnace Slag']=tugas1['Blast Furnace Slag'].fillna((tugas1['Blast Furnace Slag'].mean()))
tugas1['Fly Ash']=tugas1['Fly Ash'].fillna((tugas1['Fly Ash'].mean()))
tugas1['Superplastic']=tugas1['Superplastic'].fillna((tugas1['Superplastic'].mean()))
np.sum(tugas1.isnull())
```

```python
#normalisasi
from sklearn.preprocessing import normalize
tugas1norm=normalize(tugas1,norm='l2')
tugas1norm=pd.DataFrame(tugas1norm)
tugas1norm.tail()
```

```python
#3sigmoid
tugas1sigmoidal=(1-2.718281828**(-tugas1norm))/(1+2.718281828**(-tugas1norm))
tugas1sigmoidal=pd.DataFrame(tugas1sigmoidal)
tugas1sigmoidal.tail()
```

```python
tugas1sigmoidal.to_csv('datamiss_trans.csv', sep=',')
```
