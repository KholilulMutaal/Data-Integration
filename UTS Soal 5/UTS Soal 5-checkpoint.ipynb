```python
import pandas as pd
import numpy as np
from apyori import apriori
soal5=pd.read_csv("D:\dataset2\online_retail.csv")
soal5.head()
```

```python
np.sum(soal5.isnull())
```

```python
for i, j in soal5[soal5['Description'].isnull()].iterrows(): soal5['Description'][i]="Code-"+str(j['StockCode'])
```

```python
np.sum(soal5.isnull())
```

```python
x=soal5.groupby('InvoiceNo')
beli=[]
for i,j in x:
    beli.append(list(j['Description'].map(str)))
```

```python
data=pd.DataFrame(beli)
```

```python
data
```

```python
association_rules=apriori(beli,min_support=0.0045,
                         min_confidence=0.2,
                         min_lift=3, min_length=2)
association_result=list(association_rules)
```

```python
association_result[0]
```

```python
for item in association_result :
    #first index of the inner list
    #contains base item and add item
    pair = item[0]
    items = [x for x in pair]
    print("rule: " + items [0] + " ->" + items[1])
    
    #second ondex of the inner list
    print("support: " + str(item[1]))
    
    #third index of the list located at 10th
    #of the third index of the inner list
    
    print("cofidence: " + str(item[2][0][2]))
    print("lift: " + str(item[2][0][3]))
    print("=======================================")
```
